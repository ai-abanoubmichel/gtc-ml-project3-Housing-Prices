```python
#encoding the categorical columns 
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
#Scale range of large dataset 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#scale the large range of dataset columns
Housing_Prices[['total_rooms' ,'total_bedrooms' ,'population' ,'households']] = scaler.fit_transform(Housing_Prices[['total_rooms' ,'total_bedrooms' ,'population' ,'households']])
Housing_Prices[['median_house_value']] = scaler.fit_transform(Housing_Prices[['median_house_value']])
```

```python
#split data into features and label
features = Housing_Prices.drop('median_house_value' , axis=1)
label = Housing_Prices['median_house_value']
```

```python
#the first 5 rows of features 
features.head()
```

```python
#the first 5 rows of label
label.head()
```

```python
#split features into train & test and label into train & test 
from sklearn.model_selection import train_test_split
features_train , features_test , label_train , label_test = train_test_split(features , label , test_size=0.2 , random_state= 42 )
```

```python
#number of columns and row in features_train and label_train 
print(features_train.shape)
print(label_train.shape)
```

```python
#number of columns and row in features_test and label_test 
print(features_test.shape)
print(label_test.shape)
```

```python
#Import machine learning regression models
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR 
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor , AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor 
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
import lightgbm as lgb 

from sklearn.metrics import mean_absolute_error , r2_score 
```

```python
# Define a dictionary of regression models to train and evaluate
models = {
    'Linear_Regression Model' : LinearRegression(),
    'Decision_Tree Model' : DecisionTreeRegressor(),
    'SVR Model' : SVR(),
    'Random_Forest Model' : RandomForestRegressor() ,
    'KNeighbors_Model' : KNeighborsRegressor(),
    'AdaBoost Model' : AdaBoostRegressor(),
    'CatBoost Model' : CatBoostRegressor(),
    'XGB Model' : XGBRegressor(),
    'LighGBM Model' : lgb.LGBMRegressor()
}
```

```python
#Train the model, generate predictions, and evaluatethe model 
train_score = []
test_score = []
def impute_model(model) :
    model.fit(features_train , label_train)
    train_pred = model.predict(features_train)
    test_pred = model.predict(features_test) 

    print('=>Train metrics : ')
    print('     mean_absolute_error : ' , mean_absolute_error(label_train ,train_pred ))
    print('     r2_score : ', r2_score(label_train , train_pred))
    train_score.append(r2_score(label_train , train_pred))
    print('')

    print('=>Test Metrics : ')
    print('     mean_absolute_error : ' , mean_absolute_error(label_test , test_pred))
    print('     r2_score  : ' ,r2_score(label_test , test_pred))
    test_score.append(r2_score(label_test , test_pred))
```

```python
model = []
for model_name , model_algo in models.items() :
    print(f'=============== {model_name} ===============')
    model.append(model_name)
    impute_model(model_algo)
    print('')
```

```python
#list of train and test each model
scores = pd.DataFrame({
    'model' : model ,
    'train_score' : train_score ,
    'test_score' : test_score
})

scores
```

```python
#hyperparammeters tuning with gridsearch
from sklearn.model_selection import GridSearchCV

model = CatBoostRegressor()

params_grid = {
    'iterations': [100, 300, 500],
    'depth': [4, 6, 8, 10],
    'learning_rate': [0.01, 0.05, 0.1],
}

grid_Search = GridSearchCV(model , params_grid , cv=5 , scoring='r2' , n_jobs=-1)
grid_Search.fit(features_train , label_train)

print('========================')
print('==> Best parameters = ' , grid_Search.best_params_)
print('==> Best score = ' , grid_Search.best_score_)
```

```python
#finally train the best model with the best parameters 
model = CatBoostRegressor(depth= 8 , iterations=500 ,learning_rate=0.1)
model.fit(features_train , label_train)
```

```python
#make prediction
model_predict = model.predict(features_test)
model_predict
```
